```python
#Fianl Draft: Should only have to run this
import pandas as ps

#Imports all the data and names the coluns
data= ps.read_csv('/home/cguccione/Downloads/A_HomSap_First50Lines.csv', dtype={0:object, 1:int, 2:int, 3:object, 4:object, 5:object}, header=None)
data.columns=['Contig', 'Split', 'Gap', 'Long Name', 'Type', 'Locus', 'Gene']

#Fixes the Gene row
data['Gene'].fillna(0, inplace=True)
##NEED to try this on another data set 
#that acutally has genes in final column

#Fixes the Contig Row
clean_data=[]
for index,row in data.iterrows():
    #Gets just the numbers to work with out of
    #the SISRS_contig and names them clean
    head,tail= (row[0]).split("-")
    clean=tail.replace("/","")
    clean_data.append(clean)
data['Contig']=clean_data

#Fixes the Type Row
#Replaces each of the words with a number
data=data.replace(['Noncoding', 'Intergenic', 'CDS'],[0,1,2])

#Fixing the Long Name Row
cleanLongName=[]
longNameD={}
count=0;
for index,row in data.iterrows():
    if row[3] not in longNameD.keys():
        longNameD[row[3]]=count;
        count=count+1
    cleanLongName.append(longNameD[row[3]])
data['Long Name']=cleanLongName

#Fixing the Locus Row
cleanLocus=[]
LocusD={}
countL=0;
for index,row in data.iterrows():
    if row[5] not in LocusD.keys():
        LocusD[row[5]]=countL;
        countL=countL+1
    cleanLocus.append(LocusD[row[5]])
data['Locus']=cleanLocus

data.head()
```
